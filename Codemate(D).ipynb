```python
# Dictionary to store roll numbers and their subjects' marks
marksheet = {}

# Function to add marks for a roll number and subject
def add_marks_one_subject():
    roll_no = input("Enter your roll number: ")
    subject = input("Enter your subject: ")
    marks = float(input("Enter marks for {}: ".format(subject)))
    marksheet.setdefault(roll_no, {})[subject] = marks
    print("Marks added successfully!")

# Function to add a marksheet for a roll number
def add_marks_marksheet():
    roll_no = input("Enter your roll number: ")
    marksheet[roll_no] = {}
    subjects = int(input("Enter the number of subjects: "))
    for i in range(subjects):
        subject = input("Enter subject {}: ".format(i+1))
        marks = float(input("Enter marks for {}: ".format(subject)))
        marksheet[roll_no][subject] = marks
    print("Marksheet added successfully!")

# Function to display all marksheets
def display_marksheets():
    for roll_no, marks in marksheet.items():
        print("Roll Number:", roll_no)
        for subject, marks in marks.items():
            print(subject, ":", marks)
        print()

# Main function
def main():
    while True:
        print("\nOptions:")
        print("1. Add marks for one roll number and one subject")
        print("2. Add marks for one roll number (Marksheet)")
        print("3. Display all marksheets")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_marks_one_subject()
        elif choice == '2':
            add_marks_marksheet()
        elif choice == '3':
            display_marksheets()
        elif choice == '4':
            print("Exiting...")
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```
